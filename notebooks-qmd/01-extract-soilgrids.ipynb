# Libraries and configuration


```python
# Libraries for data extraction
import rasterio
from rasterio.plot import show
from owslib.wcs import WebCoverageService

# Custom module
import sys
sys.path.append("../source/py/")
from functions_py import extractSoilGrids
```

# Bounding boxes

- The bounding boxes for each analyzed country are set to a list, in this case: Japan, United States, Canada, Switzerland, and South Korea.
- The limits for each country were taken from the following resource on [Github.](https://gist.github.com/graydon/11198540)
- Also useful was the [bboxfinder.com](http://bboxfinder.com/#0.000000,0.000000,0.000000,0.000000)

```python
limits_japan = (129.408463169, 31.0295791692, 145.543137242, 45.5514834662)
limits_switzerland = (6.02260949059, 45.7769477403, 10.4427014502, 47.8308275417)
limits_southK = (126.117397903, 34.3900458847, 129.468304478, 38.6122429469)
limits_canada = (-140.99778, 41.6751050889, -52.6480987209, 83.23324)
limits_usa = (-131.308594, 24.686952, -63.457031, 51.289406)
```

# Extraction by country

- The following are the ten soil variables obtained from [SoilGrids.](https://www.isric.org/explore/soilgrids) In all of them the averages for the depth of 0-5 cm were extracted, except in the variable *ocs* which is only available for depth 30 cm.
    - **`bdod`:** Bulk density
    - **`cec`:** Cation exchange capacity at ph 7
    - **`cfvo`:** Coarse fragments volumetric
    - **`clay`:** Clay content
    - **`nitrogen`:** Nitrogen
    - **`phh2o`:** Soil pH in H2O
    - **`sand`:** Sand content
    - **`silt`:** Silt content
    - **`soc`:** Soil organic carbon content
    - **`ocd`:** Organic carbon densities
    - **`ocs`:** Soil organic carbon stock (depth of 30cm)

```python
# Inputs
variables = ["bdod", "cec", "cfvo", "clay", "nitrogen", "phh2o", "sand",
             "silt", "soc", "ocd", "ocs"]
resolution = 0.05
path_export = "../external-data/"
```

## Japan

```python
extractSoilGrids(variables = variables,
                 limits = limits_japan,
                 res = resolution,
                 pathExp = path_export,
                 country = "japan")
```

- Below is an example map for the variable "bdod_0-5cm_mean".

```python
map_soilgrids = rasterio.open("../external-data/bdod_0-5cm_mean_japan.tif")
show(map_soilgrids)
map_soilgrids.close()
```
